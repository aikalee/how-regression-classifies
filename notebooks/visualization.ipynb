```python
from model import SimpleNN
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def plot_boundary(model):
    xx, yy = np.meshgrid(
        np.linspace(-4, 4, 400),
        np.linspace(-4, 4, 400)
    )
    grid = np.c_[xx.ravel(), yy.ravel()]
    logits = model.forward(grid)
    labels = logits.argmax(axis=1).reshape(xx.shape)

    plt.figure(figsize=(6,6))
    plt.contourf(xx, yy, labels, alpha=0.4, cmap="Pastel1")
    # plt.scatter(X[:,0], X[:,1], c=y, s=10, cmap="Set1")
    plt.title("Multi-Head Regression NN Classifier")
    plt.show()
```

```python
checkpoint = np.load(os.path.join(PROJECT_ROOT, "outputs", "checkpoints", "2d_epoch_100.npz"))
params = {k: checkpoint[k] for k in checkpoint.files}
model = SimpleNN()
model.set_params(params)
plot_boundary(model)
```

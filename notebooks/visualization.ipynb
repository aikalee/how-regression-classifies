```python
# from model import SimpleNN
import matplotlib.pyplot as plt
import numpy as np
from dataset import make_dataset
```

```python
logits = np.load(os.path.join(PROJECT_ROOT, "outputs", "logs", "logged_logits.npy"))
x = make_dataset(seed=42)[0][:, 0]
idx = np.argsort(x) 
sorted_x = x[idx]
sorted_y0 = logits[:, 0][idx]
sorted_y1 = logits[:, 1][idx]
sorted_y2 = logits[:, 2][idx]
```

```python
def plot_boundary(x, y0, y1, y2):
    plt.figure()
    plt.plot(x, y0, label="Class 0")
    plt.plot(x, y1, label="Class 1")
    plt.plot(x, y2, label="Class 2")
    plt.show()
```
